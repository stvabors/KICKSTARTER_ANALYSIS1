```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url =" https://en.wikipedia.org/wiki/Metaverse"
req = requests.get(url)
print (req)
```

```python
req.content
```

```python
soup = BeautifulSoup( req.text, "html.parser")
```

```python
extracted_text = ""
for para in soup.find_all('p')[1:5]: ## we want to extract 2nd p tag till 5th p tag
    extracted_text = extracted_text + "\n" + para.text
```

```python

print (extracted_text)
```

# Text Summarisation

Creating a summary of long essay/paragraph/group of words is known as summarisation. There are 2 techniqu's in text summarization. 
1. `Extractive Text Summarization`  : Traditional/Old/Classifical Technique. Identify significant sentences of text and add them to summary. 

2. `Abstractive Text Summarization` : Advanced concept. Exact words/sentences are not extracted in this method. Important Sections are identifies and their context is interpretted and reproduced in a new way. 


Basic Text Summarization

```python
import gensim
from gensim.summarization import summarize
```

```python
shorSummary= summarize(extracted_text)
print( shorSummary )
```

```python
shorSummary = summarize( extracted_text, ratio = 0.6 )
print (shorSummary)
```

## LexRank Alogrithm

This is a text summarization technique.

```python
!pip install sumy
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
from sumy.summarizers.lex_rank import LexRankSummarizer
```

```python
import nltk
nltk.download('punkt')
  
```

```python
parser = PlaintextParser.from_string( extracted_text, Tokenizer('english')  )
```

```python
lexRank_summarizer = LexRankSummarizer()

shorSummary  = lexRank_summarizer( parser.document, sentences_count = 4)
```

```python
for sent in shorSummary:
    print (sent)
```
